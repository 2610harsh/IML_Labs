```python
import numpy as np
```

**1) To find the maximum and minimum value of a given flattened array.**

```python
array = np.arange(10)
print("Main array : ",array);
print("Max value : ",array.max())
print("Min value : ",array.min())
```

**2) To compute the mean, standard deviation, and variance of a given array along the second axis.**

```python
array2 = np.arange(9).reshape(3,3)

mean_axis_1 = np.mean(array2, axis=1)

std_dev_axis_1 = np.std(array2, axis=1)

variance_axis_1 = np.var(array2, axis=1)

print("Main array : \n",array2)
print("Mean along axis 1:", mean_axis_1)
print("Standard Deviation along axis 1:", std_dev_axis_1)
print("Variance along axis 1:", variance_axis_1)
```
