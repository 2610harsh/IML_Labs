```python
import numpy as np
```

**1) to split an array of 14 element into 3 array,each with 2,4 and 8 element in the orignal order**

```python
orignal_array = np.arange(14)
split_array = np.split(orignal_array,[2,6])

for i,arr in enumerate (split_array,1):
    print(f"Split Array(1):",arr)
```

**2) to stack array horizontally(column wise)**

```python
array1 = np.arange(2)
array2 = np.arange(2,6)
array3 = np.arange(2,14)

st_array = np.hstack((array1,array2,array3))
print("Stack Array:",st_array)
```
